```python
# visualization with pandas dataFrame
ice_cream_sales=pd.DataFrame({"temp":[30,43,55,43,23,31],"sales":[120,400,600,450,100,300]})
ice_cream_sales
```

```python
x=np.array(ice_cream_sales["temp"])
y=np.array(ice_cream_sales["sales"])
```

```python
fig=plt.figure(figsize=(10,5),dpi=100)
axes=fig.add_axes([0.1,0.1,0.9,0.9])
axes.set_xlabel("temp")
axes.set_label("sales")
axes.set_title("fig 2")
axes.plot(x,y,label="blue",linewidth=1,markersize=7,marker="o")
axes.legend(loc=4) # upper_right loc=1,upper_left loc=0......
axes.grid(b=True)
axes.annotate("point",xy=(30,110))
# plt.close() #clear memory
```

```python
#How to write formulas inside a figure
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.9,1])
axes.plot(x,y)
axes.text(25,500, r'$ \beta \plus \alpha $')
```
